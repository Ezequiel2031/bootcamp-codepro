```python
palabra_usuario = input('Ingrese una palabra')
numero_usuario = int(input('Ingrese un numero'))

def cantidad_letras(word, number):

    numero_de_letras = len(word)

    if numero_de_letras == number:
        return True

    else:
        return False

resultado = cantidad_letras(palabra_usuario, numero_usuario)
print (resultado)
```
